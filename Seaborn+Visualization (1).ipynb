```python
import seaborn as sns
```

```python
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
import numpy as np
```

```python
%matplotlib inline
```

```python
tips = sns.load_dataset("tips")
```

```python
tips.head(4)
```

```python
#Distribution Plots
```

```python
sns.distplot(tips["total_bill"], kde= False, bins = 30)
```

```python
sns.jointplot(x = "total_bill", y = "tip", data = tips, kind= "kde")
```

```python
sns.pairplot(tips, hue= "sex", palette = "coolwarm")
```

```python
sns.rugplot(tips["total_bill"])
```

```python
#Categorical Plots
```

```python
sns.barplot(x ="sex" , y="total_bill" , data = tips, estimator= np.std)
```

```python
sns.countplot(x = "sex", data = tips)
```

```python
sns.boxplot(x= "day",y="total_bill", data = tips, hue = "smoker")
```

```python
sns.violinplot(x = "day", y = "total_bill", data = tips, hue = "sex",split = True)
```

```python
sns.stripplot(x = "day", y = "total_bill", data = tips, jitter = True, hue = "sex", split= True)
```

```python
sns.violinplot(x = "day", y ="total_bill", data = tips)
sns.swarmplot(x = "day", y ="total_bill", data = tips, color = "black")
```

```python
sns.factorplot(x = "day", y = "total_bill", data = tips, kind = "bar")
```

```python
#Matrix Plots
```

```python
tc = tips.corr()
```

```python
sns.heatmap(tc, annot = True, cmap = "magma", linecolor = "red", linewidths = 3)
```

```python
sns.clustermap(tc, cmap = "coolwarm", standard_scale = 1)
```

```python
#Regression Plots
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, hue = "sex", markers = ['o', 'v'])
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips, hue = "sex", markers = ['o', 'v'], scatter_kws = {'s':200})
```

```python
sns.lmplot(x = "total_bill", y = "tip", data = tips,col = 'day', hue = "sex", aspect = 0.6, size = 5)
```
